```python
from ultralytics import YOLO
import torch

print("Torch:", torch.__version__)
```

```python
model = YOLO("/kaggle/input/models/rupankarmajumdar/yolo-last/pytorch/default/1/last.pt")
```

```python
model.train(
    data="/kaggle/input/datasets/rupankarmajumdar/sentinelblue/data.yaml",
    epochs=50,
    imgsz=640,
    batch=192,
    device=[0,1],
    workers=12,
    project="training_resume",
    name="yolo11_resume"
)
```
